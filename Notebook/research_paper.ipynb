```python
import pandas as pd

# Load dataset
data = pd.read_csv('D:/College/sem 7/final year project/gith/AniSearch/anime.csv')  

# Ensure 'Tags' column has no NaN values
data['Tags'] = data['Tags'].fillna('')

# Convert 'Rating Score' to numeric, handling errors
data['Rating Score'] = pd.to_numeric(data['Rating Score'], errors='coerce')

def recommend_by_genre(genre, top_n=10):
    """
    Filters anime that contain the selected genre in their 'Tags' column.
    Sorts the results by 'Rating Score' in descending order.
    """
    # Find anime that contain the selected genre as part of their tags
    selected_anime = data[data['Tags'].str.contains(rf'\b{genre}\b', case=False, na=False)]

    if selected_anime.empty:
        return None, f"No anime found with the tag: {genre}"

    # Select relevant columns
    recommended_anime = selected_anime[['Name', 'Tags', 'Rating Score', 'Number Votes']].copy()

    # Sort by Rating Score (highest to lowest)
    recommended_anime = recommended_anime.sort_values(by='Rating Score', ascending=False)

    return recommended_anime.head(top_n), None

# Define the target genre
selected_tag = "Action"

# Get top anime for the selected genre
recommended_anime, error_message = recommend_by_genre(selected_tag)

# Print results
if recommended_anime is None:
    print(error_message)
else:
    print(f"\nTop Anime for Genre '{selected_tag}' (Sorted by Rating Score):\n")
    
    # Print formatted output
    for i, (name, tags, rating, votes) in enumerate(
        zip(recommended_anime['Name'], recommended_anime['Tags'], recommended_anime['Rating Score'], recommended_anime['Number Votes']), 1
    ):
        print(f"{i}. {name.ljust(50)}  Rating: {rating:.3f}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('D:/College/sem 7/final year project/gith/AniSearch/anime.csv') 

data['Tags'] = data['Tags'].fillna('') 
data['Tags'] = data['Tags'].apply(lambda x: ', '.join(sorted(x.split(', ')))) 

data['Rating Score'] = pd.to_numeric(data['Rating Score'], errors='coerce')

def recommend_by_genre(genre, top_n=15):
    selected_anime = data[data['Tags'].str.contains(genre, case=False, na=False)]

    if selected_anime.empty:
        return None, f"No anime found with the tag: {genre}"

    recommended_anime = selected_anime[['Name', 'Tags', 'Rating Score', 'Number Votes']].copy()
    recommended_anime = recommended_anime.sort_values(by='Rating Score', ascending=False)

    return recommended_anime.head(top_n), None

selected_tag = "Action" 

recommended_anime, error_message = recommend_by_genre(selected_tag)

if recommended_anime is None:
    print(error_message)
else:
    print("\nRecommended anime sorted by rating (Top 15):")
    print(recommended_anime)

    top_n = 10  
    recommended_anime_top_n = recommended_anime.head(top_n)

    if not recommended_anime_top_n.empty:
        plt.figure(figsize=(12, 8))

        ax = sns.barplot(
            data=recommended_anime_top_n,
            x='Rating Score',
            y='Name',
            palette='plasma'
        )
        
        
        for bar, (rating, votes) in zip(ax.patches, zip(recommended_anime_top_n['Rating Score'], recommended_anime_top_n['Number Votes'])):
            bar_width = bar.get_width()  
            bar_height = bar.get_height()  
            ax.text(
                bar_width - 0.1,          
                bar.get_y() + bar_height / 2,  
                f"{rating:.1f}\n{int(votes)}",  
                va='center', ha='right', color='white', fontsize=10, weight='bold'
            )

      
        plt.xlabel('Rating Score', fontsize=12)
        plt.ylabel('Anime Name', fontsize=12)
        plt.title(f"Top {top_n} Anime for Genre '{selected_tag}' (Rating Score and Votes)", fontsize=12)

       
        min_score = recommended_anime_top_n['Rating Score'].min() - 0.2
        max_score = recommended_anime_top_n['Rating Score'].max() + 0.2
        plt.xlim(min_score, max_score)

        plt.tight_layout()
        plt.show()
    else:
        print(f"No data to plot for the genre '{selected_tag}'.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_distribution(df, numeric_columns, categorical_columns):
    
    for column in numeric_columns:
        plt.figure(figsize=(10, 6))
        sns.histplot(df[column], kde=True, bins=30, color="blue")
        plt.title(f"Distribution of {column}")
        plt.xlabel(column)
        plt.ylabel("Frequency")
        plt.show()

    for column in categorical_columns:
        plt.figure(figsize=(10, 6))
        sns.countplot(data=df, y=column, order=df[column].value_counts().index, palette="viridis")
        plt.title(f"Distribution of {column}")
        plt.xlabel("Count")
        plt.ylabel(column)
        plt.show()

numeric_columns = ['Rating Score', 'Number Votes', 'Episodes', 'Duration']
categorical_columns = ['Type', 'Season', 'Studios']

plot_distribution(df, numeric_columns, categorical_columns)
```
